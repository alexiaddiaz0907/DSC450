# **Data Cleaning ** 

```python
import string

def remove_punctuation(text):
  no_punct=[words for words in text if words not in string.punctuation]
  words_wo_punct = ''.join(no_punct)
  return words_wo_punct


google_reviews_1['reviews_wo_punct'] = google_reviews_1['summary'].apply(lambda x: remove_punctuation(x))
google_reviews_1['pros_wo_ punct'] = google_reviews_1['pros'].apply(lambda x: remove_punctuation(x))
google_reviews_1['cons_wo_punct'] = google_reviews_1['cons'].apply(lambda x: remove_punctuation(x))
google_reviews_1.head()
```

```python


def tokenize(text):
    split=re.split("\W+",text) 
    return split
google_reviews_1['reviews_wo_punct_split'] = google_reviews_1['reviews_wo_punct'].apply(lambda x: tokenize(x.lower()))
#google_reviews_1['pros_wo_punct_split'] = google_reviews_1['pros_wo_punct'].apply(lambda x: tokenize(x.lower()))
#google_reviews_1['cons_wo_punct_split'] = google_reviews_1['cons_wo_punct'].apply(lambda x: tokenize(x.lower()))

google_reviews_1.head
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')
stopwords.words()

stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
  text = [word for word in text if word not in stopword]
  return text

google_reviews_1['reviews_wo_punct_split_wo_stopwords'] = google_reviews_1['reviews_wo_punct_split'].apply(lambda x: remove_stopwords(x))
google_reviews_1.head()

```

# **Sentiment Analysis**

```python
# LDA Topic Modeling 

google_reviews_1_data = google_reviews_1['reviews_wo_punct_split_wo_stopwords']
google_reviews_1_data.head()

id2word = corpora.Dictionary(google_reviews_1_data)
texts = google_reviews_1_data

corpus = [id2word.doc2bow(word) for word in texts]
print(corpus)
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus = corpus, id2word = id2word, num_topics = 10, random_state = 100, update_every = 1, chunksize = 100, passes = 10, alpha = "auto")
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

google_reviews_1['scores'] = google_reviews_1['reviews_wo_punct_split_wo_stopwords'].apply(lambda reviews_wo_punct_split_wo_stopwords: sia.polarity_scores(str(reviews_wo_punct_split_wo_stopwords)))
print(google_reviews_1['scores'])
```

```python
google_reviews_1['compound'] = google_reviews_1['scores'].apply(lambda score_dict:score_dict['compound'])
google_reviews_1.head()
```

```python
google_reviews_1['pos'] = google_reviews_1['scores'].apply(lambda pos_dict:pos_dict['pos'])
google_reviews_1.head
```

```python
google_reviews_1['neg'] = google_reviews_1['scores'].apply(lambda neg_dict:neg_dict['neg'])
google_reviews_1.head
```

```python
google_reviews_1['type']=''
google_reviews_1.loc[google_reviews_1.compound>0, 'type'] = 'POS'
google_reviews_1.loc[google_reviews_1.compound==0, 'type'] = 'NEUTRAL'
google_reviews_1.loc[google_reviews_1.compound<0, 'type'] = 'NEG'
google_reviews_1.head()
```

```python
len = google_reviews_1.shape
(rows,cols) = len
pos = 0
neg = 0
neutral = 0
for i in range(0,rows):
  if google_reviews_1.loc[i][12]=='POS':
    pos=pos+1
  if google_reviews_1.loc[i][12]=="NEG":
    neg=neg+1
  if google_reviews_1.loc[i][12]=="NEUTRAL":
    neutral=neutral+1
print("Positive :"+str(pos) + "Negative :" + str(neg) + " Neutral:"+ str(neutral))
```
